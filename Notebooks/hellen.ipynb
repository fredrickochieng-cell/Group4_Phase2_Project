## Merging the rest of the pandas dataframes with the primary dataframe ie IMDB dataframe
In this step we are supposed to merge the main dataframe to the other dataframes i.e box office movies, The movie database and the Numbers to come up with one comprehensive dataset to help clearly analyze the patterns in the movie industry.

```python
# checking the columns on both data frames
print(df_tm_cleaned.columns)
print(merged_df.columns)
```

```python
#changing the data type of start year from an integer to string
merged_df['start_year'] = merged_df['start_year'].astype(str)
```

```python
# Changing release year to release date for merging purposes
df_tm_cleaned['release_year'] = pd.to_datetime(df_tm_cleaned['release_date'], errors='coerce').dt.year
```
