We fill the null values in the column with missing values.

```python
#fill the null values in the column with missing values.
#In this case, we are filling the null values in the 'critic' column with the string 'Missing' because its cate
df_reviews['critic'].fillna('Missing', inplace=True)
```

```python
#We fill rating column null values with 'Not Rated'
df_reviews['rating'].fillna('Not Rated', inplace=True)
```

```python
#fill the review column null values with 'No Review' since it is textual data and mode would not quite work since its creative writing
df_reviews['review'].fillna('No Review', inplace=True)
```

```python
publisher_mode = df_reviews['publisher'].mode()
publisher_mode[0]
```

```python
#fill the piblisher column null values with the mode for the column publisher since it is categorical data
df_reviews['publisher'].fillna(publisher_mode, inplace=True)
```
